```python
import time
import sys
```

```python
sys.setrecursionlimit(2000)
```

```python
# Função recursiva para calcular o fatorial
def fatorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * fatorial(n - 1)
```

```python
# Teste da função fatorial e medição do tempo de execução
try:
    numero = int(input("Digite um número inteiro para calcular o fatorial: "))
    if numero < 0:
        print("Digite um número inteiro >= 0")
    else:
        inicio = time.perf_counter()
        resultado = fatorial(numero)
        fim = time.perf_counter()
        tempo = fim - inicio
        print(f"O fatorial de {numero} é {resultado}")
        print(f"Tempo de execução: {tempo:.6f} segundos")

except ValueError:
    print("Entrada inválida")
```

#### Tempos de Execução para valores de n:
- n = 10: 0.000003 segundos
- n = 100: 0.000021 segundos
- n = 500: 0.000154 segundos
- n = 1000: 0.000445 segundos

O algoritmo possui complexidade linear O($n$). Porque, para um valor $n$, a função é chamada recursivamente $n$ vezes, sendo assim o crescimento de 1 para 1. 

Aqui está a fórmula de complexidade da função:  $T_n = T_{n-1} + 1$
